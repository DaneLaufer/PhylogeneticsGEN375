```python
import os
import json
import pandas as pd
```

```python
def checkFullSequence(checkSeq, knownSeq):
    seqLength = len(knownSeq)
    differences = 0
    for letterIdx in range(seqLength):
        knownLetter = knownSeq[letterIdx]
        testLetter = checkSeq[letterIdx]

        if knownLetter != testLetter:
            differences += 1
    return differences
```

```python
def compareSamples(sample1, sample2):
    sample1results = json.load(open(os.path.join('results', sample1 +".json"), 'r'))
    sample2results = json.load(open(os.path.join('results', sample2 +".json"), 'r'))
    results = {}
    for geneName in sample1results:  
        try:
            sample1seq = (sample1results[geneName]['sequence'])
        except KeyError:
            sample1seq = None
        try:
            sample2seq = (sample2results[geneName]['sequence'])
        except KeyError:
            sample2seq = None

        if sample1seq and sample2seq:
            results[geneName] = checkFullSequence(sample1seq, sample2seq)
        else:
            string = 'NA'
            if not sample1seq:
                string += sample1[-1]
            if not sample2seq:
                string += sample2[-1]
            results[geneName] = string

    return results
    
```

```python
def generateDf(sequence):
    data = {}
    for file in os.listdir('results'):
        seqName = file.replace('.json', '')
        if seqName == sequence:
            continue
        data[seqName] = compareSamples(sequence, seqName)
    return pd.DataFrame.from_dict(data)
```

```python
generateDf('sequence2')
```

```python
a = pd.Series(compareSamples('sequence1', 'sequence10'), name = '10')
```
